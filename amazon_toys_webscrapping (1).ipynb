```python
import requests
from bs4 import BeautifulSoup     #Import the BeautifulSoup class from the bs4
import re
import time
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'
}
```

```python
base_url = 'https://www.amazon.in/s?k=toys&page={}'
```

```python
product_data = []
```

```python
for page in range(1, 3):
    url = base_url.format(page)
    print(f"Fetching page {page}: {url}")
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Failed to retrieve page {page}")
        continue
```

```python
    soup = BeautifulSoup(response.content, 'html.parser')
    products = soup.find_all('div', {'data-component-type': 's-search-result'})
```

```python
for product in products:
    img_elem = product.find('img')
    img_url = img_elem['src'] if img_elem and 'src' in img_elem.attrs else 'N/A'
    print(f"Image URL: {img_url}")
```

```python
#for extracting titles
for product in products:
    title_elem = product.h2
    title = title_elem.text.strip() if title_elem else 'N/A'
    print(f"Title: {title}")
    print("-" * 40)
```

```python
#for extracting ratings
for product in products:
    rating_elem = product.find('span', class_='a-icon-alt')
    rating = rating_elem.get_text(strip=True) if rating_elem else 'N/A'
    print(f"Rating: {rating}")
    print("-" * 40)
```

```python
#for extracting prices
for product in products:
        price_elem = product.find('span', class_='a-price-whole')
        price = price_elem.get_text(strip=True) if price_elem else 'N/A'
        print(f"Price: ₹{price}")
        print("-" * 40)
    
```

```python
for product in products:
    title_elem = product.h2
    title = title_elem.text.strip()
    price_elem = product.find('span', class_='a-price-whole')
    rating_elem = product.find('span', class_='a-icon-alt')
    product_data.append({
            'Title': title,
            'Price': price,
            'Rating': rating
    })
    
```

```python
for product in product_data:
    print(f"{product['Title']}\t\t------>{product['Price']}------>{product['Rating']}")
```

```python
df = pd.DataFrame(product_data)
df.to_csv('amazon_toys.csv', index=False)
print("dataset saved sucessfully")
```
